```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def calcx(m, d, N):
    return d * (2 * m - N)

def calct(N, tau):
    return N * tau

def calcm(x, d, N):
    return (N + x / d) / 2

def calcN(t, tau):
    return t / tau
def P_N_m(N, m):
    # The following works but is unstable for large N
    #if m > N:
    #    return 0
    #else:
    #    return (1/2)**N * factorial(N) / (factorial(N - m) * factorial(m))
    return binom.pmf(m, N, 1/2)

def P(x, t, tau, d):
    N = int(round(t / tau))
    m = int(round((N + x /d) / 2 ))
    return binom.pmf(m, N, 1/2) / (2*d)
def normalpdf(x, t, D):
    return (4 * np.pi * D * t)**(-1/2) * np.exp(-x**2 / (4 * D * t))
P_N_m(0, 2)
np.float64(0.0)
trialN, trialm = 200, 5
trialPNm = P_N_m(trialN, trialm)

d = 0.0001
tau = 1e-2
trialx, trialt = calcx(trialm, d, trialN), calct(trialN, tau)
trialP = P(trialx, trialt, tau, d)

transformationerror = trialPNm - 2*d *trialP
assert transformationerror < 1e-9
print(f'The transformation error is small: {transformationerror}')

trialD = d**2 / (2 * tau)
trialNormP = normalpdf(trialx, trialt, trialD)
normapproxerror = trialNormP - trialP
assert normapproxerror < 1e-9
print(f'The normal approximation error is small: {normapproxerror}')
#The transformation error is small: 0.0
#The normal approximation error is small: 1.8001888863060975e-37
#To get a finer grid of points for 
# we reduce the increment 
 #but must then also increase the number of samples 
#if we want to cover the same domain for 
.

trialN = 10
M = np.arange(1, trialN)

d = 0.1
plt.scatter(M, [calcx(m, d, trialN) for m in M], label=f'd = {d}')
d = 0.01
plt.scatter(M, [calcx(m, d, trialN) for m in M], label=f'd = {d}')
plt.legend()
plt.xlabel("m")
plt.ylabel("x(m)")
plt.show()
```
