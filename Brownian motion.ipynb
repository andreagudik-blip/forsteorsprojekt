```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
def brown(mu,sigma,N):
    return np.cumsum(np.random.normal(mu,sigma,N))
```

```python
mu=0
sigma=1
N=1000

plt.plot(brown(mu,sigma,N),brown(mu,sigma,N))
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.plot(brown(mu,sigma,N),brown(mu,sigma,N),brown(mu,sigma,N))
```
